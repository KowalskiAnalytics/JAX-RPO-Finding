```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
omega = 10 * 2 * np.pi

t = np.linspace(0, 0.5, 1000)
y = np.sum([n * np.cos( n * omega * t ) for n in range(N+1)],0)

A = np.real(np.fft.rfft(y,1000))

plt.plot(t,y)
plt.show()

plt.plot(A[0:50])
plt.show()

print(A[:50])

plt.plot(np.fft.ifft(A))
plt.show()

# 500 -> 0.5, d = 1.0, and I want 25 -> 50 so d = 1/2*1000
freqs = np.fft.rfftfreq(1000,1/2000)
plt.plot(freqs)
plt.show()

Y = np.zeros(1000)
for n in range(50):
    Yn = A[n] * np.cos(t * freqs[n])
    plt.plot(Yn)
    Y += Yn
plt.show()

plt.plot(Y)
plt.show()
```
